```python
import pandas as pd
from amr import AMR
import penman
from huamr.utils.amr_helper import filter_valid_amrs
```

```python
def load_synthetic_data():
    df = pd.read_csv('/home/bart/data/amr/amr_annotation_3.0/data/amrs/synthetic/generated.csv')
    df = df.rename(columns={'generated_amr': 'amr_graph'})
    df = filter_valid_amrs(df, 'amr_graph')
    return df
```

```python
df = load_synthetic_data()
```

```python
df['parsed'] = df['amr_graph'].apply(AMR.parse_AMR_line)
```

```python
df['parsed'].apply(lambda x: x.node_values)
```

```python
df['node_values'] = df['parsed'].apply(lambda x: x.node_values)
df['top_node'] = df['node_values'].apply(lambda x: x[0])
```

```python
df['top_node'].value_counts()
```

```python
df.explode('node_values')['node_values'].value_counts()
```

```python
df['decoded'] = df['amr_graph'].apply(penman.decode)
```

```python
df['encoded'] = df['decoded'].apply(penman.encode)
```
